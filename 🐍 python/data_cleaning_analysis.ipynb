## Feature Engineering

## Calculate the Order_Value

```python
#item-level revenue
order_items['total_value'] = order_items['price'] + order_items['freight_value']
order_items['total_value']
```

## Aggregations

```python
# Expected Revenue per Order
order_value = order_items.groupby('order_id')['total_value'].sum().reset_index()
```

## Calculate Payment Value

```python
# Actual Payment Value per order
payment_value = payments.groupby('order_id')['payment_value'].sum().reset_index()
```

##  Reconciliation

## Merge Both

```python
df = order_value.merge(payment_value, on = 'order_id', how = 'inner')
df
```

```python
## Calculate Difference
df['difference'] = (df['total_value'] - df['payment_value']).round(2)
df['difference']
```

```python
## Mismatch Identification
df[df['difference'] != 0]
```

```python
df['difference'].value_counts()
```

## Segmentation

```python
# Categorize
# Clean Data
def categorize(x):
    if abs(x) <= 0.5:
        return 'Match'
    elif x > 0:
        return 'Underpaid'
    else:
        return 'Overpaid'

df['category'] = df['difference'].apply(categorize)
```

```python
df.head()
```

```python
#Count summary
df['category'].value_counts()
```
